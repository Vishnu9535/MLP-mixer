```python
import pandas as pd
data = pd.read_csv(r'C:\Users\vishn\OneDrive\Documents\tsmixer\final_dataset_nifty_bees.csv')
```

```python
columns_to_normalize = ['High Price  ','Low Price  ', 'Last Price  ', 'Close Price  ',
       'Average Price ', 'Total Traded Quantity  ','Turnover ₹  ']
for column in columns_to_normalize:
    data[column] = data[column].replace({',': ''}, regex=True)
    data[column] = data[column].astype(float)
data.head()
```

```python
data = data.rename(columns={'Date  ':'date'})
data = data.rename(columns={'Open Price  ':'Open_Price'})
data = data.rename(columns={'High Price  ':'High_Price'})
data = data.rename(columns={'Low Price  ':'Low_Price'})
data = data.rename(columns={'Total Traded Quantity  ':'Volume'})
data = data.rename(columns={'Close Price  ':'Close_Price'})
```

```python
data.head()
```

```python
columns_to_normalize = ['Open_Price', 'High_Price', 'Low_Price', 'Volume','Close_Price']  
data[columns_to_normalize].head()
```

```python
def min_max_normalize(data, columns):
    for column in columns:
        data[column] = (data[column] - data[column].min()) / (data[column].max() - data[column].min())
    return data
```

```python
data[columns_to_normalize] = data[columns_to_normalize].apply(pd.to_numeric, errors='coerce')
data = min_max_normalize(data, columns_to_normalize)
```

```python
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
data['date'] = data['date'].dt.strftime('%m/%d/%Y')
data['date'].head()
```

```python
columns_to_normalize= ['date','Open_Price', 'High_Price', 'Low_Price', 'Volume','Close_Price']
data[columns_to_normalize].to_csv('final_dataset_normalized.csv')
```
